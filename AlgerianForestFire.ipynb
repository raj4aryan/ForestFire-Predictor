## Algerian Forest Fires Dataset 
Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attribues and 1 output attribue (class)
The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

+ For Classification we take classes feature, for regression we use FWI feature

Attribute Information:

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8
FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfOrg = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv", header = 1)
df = dfOrg.copy()
```

```python
df.info()
```

### Data Cleaning

```python
# Missing Values
df[df.isnull().any(axis=1)]
```

1. The dataset is partitioned in two regions, we need to merge the dataset.
2. We can make a new column based on the region and assign labels like 0 or 1

```python
# we create a new attribute and categorise the index based on region
df.loc[:122, "Region"] = 0
df.loc[122:, "Region"] = 1
# initially region is of float type, we convert to int to save space
df[['Region']] = df[['Region']].astype(int)
df.info()
```

```python
#Removing the null data
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
#Also we need to drop the in between header
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop = True)
df.iloc[[122]]
```

Now, the data has been merged and dataset is free from null values

```python
df.columns
```

```python
# Notice, the attribute name Classes has space in its name, we remove it
df.columns = df.columns.str.strip()
df.columns
```

### Feature Engineering

```python
#Attributes from Temperature to FWI are all numerical type but they are mentioned as object/str. so we need to change it
df.head()
```

```python
df[['day', 'month',	'year',	'Temperature', 'RH',	'Ws']] = df[['day', 'month', 'year',	'Temperature', 'RH', 'Ws']].astype(int)
df.info()
```

```python
# Now we need to update the float columns
objects = [features for features in df.columns if df[features].dtype == 'str']
for feature in objects:
    if feature!='Classes':
        df[feature] = df[feature].astype(float)
```
